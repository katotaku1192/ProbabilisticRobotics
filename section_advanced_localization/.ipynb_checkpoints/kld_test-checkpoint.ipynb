```python
import sys
sys.path.append('../scripts/')
from robot import *
from scipy.stats import norm, chi2
```

```python
def num(epsilon, delta, binnum):                                        # 必要なパーティクルの数を計算
    return math.ceil(chi2.ppf(1.0 - delta, binnum - 1) / (2*epsilon))   # 端数は切り上げ
```

```python
fig, (axl, axr) = plt.subplots(ncols=2, figsize=(10, 4))

bs = np.arange(2, 10)
n = [num(0.1, 0.01, b) for b in bs]                                     # ビンの数が2から10までのパーティクル数
axl.set_title("bin: 2-10")
axl.plot(bs, n)

bs = np.arange(2, 100000)
n = [num(0.1, 0.01, b) for b in bs]                                     # ビンの数が2から100000までのパーティクル数
axr.set_title("bin: 2-100000")
axr.plot(bs, n)

plt.show()
```
