```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_600.txt", delimiter=" ", header=None, names=("date", "time", "ir", "lidar"))

data["lidar"].hist(bins = max(data["lidar"])-min(data["lidar"]), align="left")
plt.show()
```

```python
data.lidar.plot()
plt.show()
```

```python
data["hour"] = [e//10000 for e in data.time]
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist(alpha=0.3)
d.lidar.get_group(14).hist(alpha=0.3)
plt.show()
```

```python
each_hour = { i : d.lidar.get_group(i).value_counts().sort_index() for i in range(24)} #時間帯ごとにセンサ値を数え、データフレームを作る
freqs = pd.concat(each_hour, axis=1) # 連結
freqs = freqs.fillna(0)
probs = freqs / len(data) #確率に

probs
```

```python
import seaborn as sns

sns.heatmap(probs)
plt.show()
```

```python
sns.jointplot(data["hour"], data["lidar"], data, kind="kde")
plt.show()
```

```python
p_t = pd.DataFrame(probs.sum()) #各列を合計
p_t.plot()
p_t.transpose()
```

```python
p_t.sum()
```

```python
p_z = pd.DataFrame(probs.transpose().sum())
p_z.plot()
p_z.transpose()
```

```python
p_z.sum()
```

```python
cond_z_t = probs / p_t[0] #列（時間）ごとにP(t)で割るとP(x|t)になる
cond_z_t
```

```python
cond_z_t.sum()
```

```python
(cond_z_t[6]).plot.bar(color="blue", alpha=0.5)
(cond_z_t[14]).plot.bar(color="orange", alpha=0.5)
plt.show()
```

```python
cond_t_z = probs.transpose()/probs.transpose().sum() # 行と列を入れ替えて同様に計算するとP(t|z)となる

print("P(z=630) = ", p_z[0][630]) # センサ値が630になる確率（何時かの情報はない）
print("P(t=13) = ", p_t[0][13]) # 時間帯が13時である確率
print("P(t=13 | z=630) = ", cond_t_z[630][13]) # センサ値が630という条件の下で、時間帯が13時である確率

print("Bayes P(z=630 | t=13) = ", cond_t_z[630][13] * p_z[0][630] / p_t[0][13]) # 時間帯が13時のとき、センサ値が630である確率（ベイズの公式より）
print("Answer P(z=630 | t=13) = ", cond_z_t[13][630]) # ↑の答え　P(z|t)
```

```python
def bayes_estimation(sensor_value, current_estimation):
    new_estimation = []
    for i in range(24):
        new_estimation.append(cond_z_t[i][sensor_value] * current_estimation[i])
        
    return new_estimation / sum(new_estimation)
```

```python
estimation = bayes_estimation(630, p_t[0])
plt.plot(estimation)
```

```python
values_5 = [630, 632, 636]

estimation = p_t[0]
for v in values_5:
    estimation = bayes_estimation(v, estimation) # 式2.58より、前のestimationを引数にして掛け合わせて求められる

plt.plot(estimation)
```

```python
values_5 = [617, 624, 619, 618, 617, 618, 620, 616, 617, 618]

estimation = p_t[0]
for v in values_5:
    estimation = bayes_estimation(v, estimation) # 式2.58より、前のestimationを引数にして掛け合わせて求められる

plt.plot(estimation)
```
